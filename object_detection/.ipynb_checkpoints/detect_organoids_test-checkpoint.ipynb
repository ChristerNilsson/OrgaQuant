# Test Organoid Detection Function (detect_organoids)

```python
import sys
sys.path.append("..")
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
from utils import visualization_utils as vis_util
from utils import label_map_util
from organoid_utils import detect_organoids
%matplotlib inline

# Set variables
# Path to frozen detection graph. This is the actual model that is used for the object detection.
PATH_TO_CKPT = "C:/Users/Timothy/Desktop/OD/object_detection/organoid_inference_graph/frozen_inference_graph.pb"
# List of the strings that is used to add correct label for each box.
PATH_TO_LABELS = "C:/Users/Timothy/Desktop/OD/object_detection/data/organoid_label_map.pbtxt"
# Path to image of organoids
IMAGE_PATH = "C:/Users/Timothy/Desktop/OD/object_detection/test_images/image5.jpg"

# Open image and load into numpy array which is required for most of the object_detection functions
image = Image.open(IMAGE_PATH)
(im_width, im_height) = image.size
image_np = np.array(image.getdata()).reshape((im_height, im_width, 3)).astype(np.uint8)

# Run the detection
[boxes, scores, classes, num, category_index] = detect_organoids(image_np,PATH_TO_CKPT, PATH_TO_LABELS)
```

```python
# TO DO put this part in detect_organoids and add threshold as input also change the edge to use im size

thresh = 0.95
# Iterate through boxes and remove edge ones
for i, box in enumerate(boxes[0]):
    if (box[0]<0.001) or (box[1] <0.001) or (box[2]>0.99) or (box[3]>0.99):
        scores[0][i] = 0

# Filter low scores below threshold   
high_score_indices = np.reshape(np.where(np.greater(scores, thresh))[1],[-1]).astype(np.int32)
boxes_filtered = np.take(boxes,high_score_indices,axis=1)
```

```python
# Visualize the results
vis_util.visualize_boxes_and_labels_on_image_array(
    image_np,
    np.squeeze(boxes_filtered),
    np.squeeze(classes).astype(np.int32),
    np.squeeze(scores),
    category_index,
    use_normalized_coordinates=True,
    max_boxes_to_draw=None,
    min_score_thresh=0.0,
    line_thickness=1)
plt.figure(figsize=(15,15))
plt.imshow(image_np)
```
