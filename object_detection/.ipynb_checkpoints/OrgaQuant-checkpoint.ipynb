# OrgaQuant
Quantify the organoid distribution from a brightfield image. This script uses a pre-trained deep learning model to locate the organoids and then quantify both diameters and area.

## Set all variables
* PATH_TO_CKPT:  Path to frozen detection graph you want to use.
* PATH_TO_LABELS: List of the strings that is used to add correct label for each box
* IMAGE_PATH: Absolute path to image of the organoids you want to quantify
* look_pix: The size of the window, a good number is less than 600 pixels but bigger than an organoid
* slide_pix: To achieve redundancy and capture edge organoids, this is a sliding window

Note: look_pix and slide_pix determine the total number of windows for each frame and hence the total time required 

```python
PATH_TO_CKPT = "C:/Users/Timothy/Desktop/OD/object_detection/organoid_inference_graph/frozen_inference_graph.pb"
PATH_TO_LABELS = "C:/Users/Timothy/Desktop/OD/object_detection/data/organoid_label_map.pbtxt"
IMAGE_PATH = "C:/Users/Timothy/Desktop/OD/object_detection/test_images/image1.jpg"
look_pix = 500
slide_pix = 250
```

## Import libraries

```python
import sys
sys.path.append("..")
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
from utils import visualization_utils as vis_util
from organoid_utils import detect_all_organoids, get_metrics
%matplotlib inline
```

# Get bounding boxes
This takes several minutes to run. It returns organoids with a 95% confidence.

```python
[boxes, scores, classes, category_index, padded_image_np, full_image_np] = detect_all_organoids(IMAGE_PATH,PATH_TO_CKPT, 
                                                                                                PATH_TO_LABELS, 
                                                                                                NUM_CLASSES = 1, 
                                                                                                look_pix=look_pix, 
                                                                                                slide_pix=slide_pix)
```

# Visualize the detection results

```python
vis_util.visualize_boxes_and_labels_on_image_array(
    full_image_np,
    np.squeeze(boxes),
    np.squeeze(classes).astype(np.int32),
    np.squeeze(scores),
    category_index,
    use_normalized_coordinates=False,
    max_boxes_to_draw=None,
    min_score_thresh=0.00,
    line_thickness=3)

plt.figure(figsize=(18,18))
plt.imshow(full_image_np)
```
